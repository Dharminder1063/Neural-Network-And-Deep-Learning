```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in data.columns:
  if data[i].dtype == "object" or data[i].dtype == "category":
    data[i] = le.fit_transform(data[i])
```

```python
x = data.drop("Attrition_Flag",axis = 1)
```

```python
y = data["Attrition_Flag"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
model = Sequential([
    Dense(21, activation='relu', input_shape=(21,)),   # Input layer + first hidden layer
    Dropout(0.3),                                 # Prevent overfitting
    Dense(32, activation='relu'),                 # Second hidden layer
    Dense(1, activation='sigmoid')                # Output layer (binary classification)
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',  # Corrected loss function for binary classification
    metrics=['accuracy']
)
```

```python
history = model.fit(x, y, epochs=20, batch_size=8, validation_split=0.2, verbose=1)
```
