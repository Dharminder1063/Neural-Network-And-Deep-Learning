```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,2))
plt.imshow(x_train[5] )
plt.show()
```

```python
we = Sequential()
we.add(Flatten(input_shape=(28,28)))
we.add(Dense(256,activation ='relu'))
we.add(Dense(10,activation='softmax'))
```

```python
we.compile(optimizer = "adam",loss="sparse_categorical_crossentropy",metrics =["accuracy"])
```

```python

we.fit(x_train,y_train,epochs = 10)
```

```python
from sklearn.metrics import accuracy_score
import numpy as np
y_pred = we.predict(x_test)
y_pred_classes = [np.argmax(element) for element in y_pred]
accuracy_score(y_test,y_pred_classes)
```
